```python
import pandas as pd
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
#we want to see how much time each user use on social media
time_data={'Facebook':120,'youtube':90,'instagram':40,'snapchat':28} #dictionary
#convert dict to a list
social_media=list(time_data.keys())
time_spend =list(time_data.values())
```

```python
style.use('ggplot') #background
plt.figure(figsize= (5,5))
color=['red','green','purple','royalblue']
plt.bar(social_media,time_spend,color=color,width=0.5)
plt.xlabel('Social media')
plt.ylabel('Time spent in Minutes')
plt.title('Avarage time spent on social media')

```
